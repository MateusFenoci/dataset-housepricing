```python
import pandas as pd
import folium
```

```python
df = pd.read_csv("dataset.csv")
```

```python
lat = df["Lat"]
long = df["Long"]

map = folium.Map(location=[lat.mean(),long.mean()],zoom_start=10)
```

```python
for lat, lon in zip(lat, long):
    folium.Marker([lat, lon]).add_to(map)
```

```python
map.save('map.html')

map
```
